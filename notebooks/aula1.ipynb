```python


# Exemplo de dados
anos = [2010, 2011, 2012, 2013, 2014, 2015]
data = {
    'Brasil': [1, 2, 3, 4, 5, 6],
    'Colômbia': [2, 3, 4, 5, 6, 7],
    'Argentina': [3, 4, 5, 6, 7, 8],
    'Peru': [4, 5, 6, 7, 8, 9]
}
df = pd.DataFrame(data, index=anos)

# Criar uma figura
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Ajustar espaçamento
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Gráficos de barras
colors = ['blue', 'green', 'red', 'purple']

# Brasil
axs[0, 0].bar(df.index, df['Brasil'], color=colors[0])
axs[0, 0].set_title('Brasil')

# Colômbia
axs[0, 1].bar(df.index, df['Colômbia'], color=colors[1])
axs[0, 1].set_title('Colômbia')

# Argentina
axs[1, 0].bar(df.index, df['Argentina'], color=colors[2])
axs[1, 0].set_title('Argentina')

# Peru
axs[1, 1].bar(df.index, df['Peru'], color=colors[3])
axs[1, 1].set_title('Peru')

# Ajustar localizador dos eixos x
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(1))

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Exemplo de dados
anos = [2010, 2011, 2012, 2013, 2014, 2015]
data = {
    'Brasil': [1, 2, 3, 4, 5, 6],
    'Colômbia': [2, 3, 4, 5, 6, 7],
    'Argentina': [3, 4, 5, 6, 7, 8],
    'Peru': [4, 5, 6, 7, 8, 9]
}
df = pd.DataFrame(data, index=anos)

# Configurações do gráfico
fig, ax = plt.subplots(figsize=(12, 8))

# Número de países e largura das barras
n_paises = len(df.columns)
largura = 0.2  # Largura das barras
espaco = 0.05  # Espaço entre grupos

# Posições das barras
posicoes = np.arange(len(df.index))

# Plotar as barras para cada país
for i, pais in enumerate(df.columns):
    ax.bar(posicoes + i * (largura + espaco), df[pais], largura, label=pais)

# Configurações do gráfico
ax.set_xlabel('Ano')
ax.set_ylabel('Valor')
ax.set_title('Dados por Países')
ax.set_xticks(posicoes + largura * (n_paises - 1) / 2)
ax.set_xticklabels(df.index)
ax.legend()

# Exibir o gráfico
plt.show()
```
